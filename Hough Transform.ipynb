```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

Given a point $P_{0} = (x_{0}, y_{0})$ all the lines that intercept $P_{0}$ have equations of the form $y_{0} = mx_{0} + b$, if we parametrize this equation we get $b = -mx_{0} + y_{0}$ which is the equation of a line in the parameters space that contains the paramenters $m$ and $b$ of all the lines that intercepts $(x_{0},y_{0})$.


For example, given $P=(2,5)$, the line $b = -m2 + 5$ in the parameters space contains the values of $m$ and $b$ of all the lines that intercepts $(2,5)$, therefore for $m = 1$ we have $b = -1x2 + 5 = 3$, what gives us the line $y = x + 3$ in the $xy$ plane and the point $(1,3)$ in the parameter space. The plots below show some lines that intercept $P$ and their respective parametrizations.

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

x = np.linspace(-1,5,100)
ax[0].plot(x,x*1 + 3, color='tab:orange', zorder=1, label='y = x + 3')
ax[0].plot(x,x*2 + 1, color='tab:blue', zorder=1, label='y = 2x + 1')
ax[0].plot(x,x*-1 + 7, color='tab:green', zorder=1, label='y = -x + 7')
ax[0].scatter(2,5,color='gray', zorder=2)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend()

m = np.linspace(-2,3,100)
ax[1].plot(m,m*(-2) + 5,color='grey', zorder=1)
ax[1].scatter(1,3, color='tab:orange', zorder=2)
ax[1].scatter(2,1, color='tab:blue', zorder=2)
ax[1].scatter(-1,7, color='tab:green', zorder=2)
ax[1].set_xlabel('b')
ax[1].set_ylabel('m');
```

### Detecting Lines

If two points $P_{0}$ and $P_{1}$ are in the same line $r$ in the $xy$-plane, each of these points define a line in the parameter space, and the point $(m,b)$ where the two lines intercept each other represents the line $r$. Thus, we can detect lines in the $xy$-plane by finding points of interception between lines in the parameter space. 

However, this parametriztion poses a problem since the more vertical a line is, the closer to infinity its value of $m$ will be. Therefore, another parametrization is used, **polar coordinates**.

#### Quick Recap

Polar coordinates are a system of coordinates where the position of a point is given by its distance $r$ to a reference point and a angle $\theta$ from a reference direction. Here, our reference point will be the origin ant the reference direction will be the $x$-axis, with $\theta$ measured counterclockwise from the axis.

The cartesian coordinates $(x,y)$ can be converted to polar coordinates using the following formulas:

\begin{cases}
x = rcos(\theta)\\
y = rsin(\theta)
\end{cases}

From this we can derive the following equation of the line in polar coordinates:

$r = xcos(\theta) + ysin(\theta)$

And similarly to the case in the paramenter space where given a point $P_{0} = (x_{0}, y_{0})$ in the cartesian plane the set of all the lines that intercept this point is given by the points $(m, b)$ in the line $b = -mx_{0} + y_{0}$. 

For polar coordinates, given a point $P_{0}$ the set of all the lines that intercept this point is given by the points $(r,\theta)$ in the sine wave $r = x_{0}cos(\theta) + y_{0}sin(\theta)$..

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

x = np.linspace(-1,5,100)
ax[0].plot(x,x*1 + 3, color='tab:orange', zorder=1, label='y = x + 3')
ax[0].plot(x,x*2 + 1, color='tab:blue', zorder=1, label='y = 2x + 1')
ax[0].plot(x,x*-1 + 7, color='tab:green', zorder=1, label='y = -x + 7')
ax[0].scatter(2,5,color='gray', zorder=2)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend()

theta = np.linspace(-2*np.pi,np.pi,100)
ax[1].plot(theta,2*np.cos(theta) + 5*np.sin(theta),color='grey', zorder=1)
ax[1].scatter(np.pi/4,-(7*(np.sqrt(2) - 1))/(np.sqrt(2) - 2), color='tab:green', zorder=2) 
ax[1].scatter(2*(np.arctan(2 - np.sqrt(5))),(np.sqrt(5) - 2)/(2*np.sqrt(5) - 5), color='tab:blue', zorder=2) 
ax[1].scatter(-np.pi/4,(3*(np.sqrt(2) - 1))/(np.sqrt(2) - 2), color='tab:orange', zorder=2)
ax[1].set_xlabel('theta')
ax[1].set_ylabel('r');
```
